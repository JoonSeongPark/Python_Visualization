# Seaborn 이란?
 : matplotlib 기반의 다양한 색 테마, 차트 기능을 추가한 python 시각화 라이브러리   
 matplotlib에 없는 그래프(hitmap, countplot ...)를 가지고 있다.  
 <br>
 <b>seaborn</b>은..  
 <br>
 다양한 그래프를 <b>간단한 code</b>를 통해 구현이 가능하고,  
 보다 <b>다채로운 색채</b>를 지원해준다.  
 
 ## Seaborn Basic Steps
 1. Data Preparing
 2. Figure Aesthetics
 3. Seaborn Plotting
 4. Further Customizations
 5. Show or Save Plot


```python
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips") # 팁 데이터 불러오기
sns.set_style("whitegrid")

# lmplot은 변수 사이의 상관관계를 파악할 때 사용.
g = sns.lmplot(x = "tip", y = "total_bill", data = tips, aspect = 2)
g = (g.set_axis_labels("Tip","Total bill(U$D)"), g.set(xlim = (0,10), ylim = (0,100)))
plt.title("Relation between Bill & Tip")

plt.show(g)
```

# 2. Figure Aesthetics
 : figure의 특성을 세팅하기!
### ※ Subplot ???
 : 한 화면에 여러 그래프를 나눠서 그려주는 기능으로 <b>plt.subplot(###)</b> 형태로 사용한다.  
<br>
subplot 안에 3개의 숫자가 들어가는데,  
앞의 두 숫자는 <b>'(# x #) 형태로 화면을 분할한다'</b>는 것을 의미한다.  
마지막 숫자는 앞의 (# x #)형태의 <b>'화면 분할 중에 # 번째 칸'</b>을 의미한다.  
<br>
<b>예시를 통해 이해해보자!</b>

```python
plt.figure(figsize = (10,6))

sns.set() # 윗 내용에서 세팅한거 초기화

plt.subplot(221) # (2x2) 4칸으로 나눴을 때 1번 칸
plt.subplot(222) # (2x2) 4칸으로 나눴을 때 2번 칸
plt.subplot(212) # (2x2) 4칸으로 나눴을 때 2번 칸

plt.show()
```

<b>이해를 위해 조금 더 복잡한 예시를 확인해봅시다!</b>

```python
plt.figure(figsize = (10,12))

plt.subplot(411)
plt.subplot(423)
plt.subplot(424)
plt.subplot(437)
plt.subplot(438)
plt.subplot(439)
plt.subplot(414)

plt.show()
```

<b>그렇다면, 각 plot에 그래프를 넣어보자!</b>  

```python
import numpy as np

sns.set_style("whitegrid")

arange = np.arange(0.01, np.pi, 0.01)

plt.figure(figsize = (10, 12))

plt.subplot(411)
plt.plot(arange, arange, 'r')
plt.grid()

plt.subplot(423)
plt.plot(arange, arange**2, 'k')
plt.grid()

plt.subplot(424)
plt.plot(arange, np.sqrt(arange), 'k')
plt.grid()

plt.subplot(437)
plt.plot(arange, np.sin(arange))
plt.grid()

plt.subplot(438)
plt.plot(arange, np.cos(arange))
plt.grid()

plt.subplot(439)
plt.plot(arange, np.tan(arange))
plt.grid()

plt.subplot(414)
plt.plot(arange, np.log(arange), 'orange')
plt.grid()

plt.show()
```

### ※ Subplot's' ???
 : plt.subplot과 plt.subplots는 사용하려는 기능은 같지만 조금 다르게 작용한다.  
 <b>'s'</b>가 붙으면 무엇이 달라지는지 살펴보자.  
 <br>
 <b>plt.subplots</b> (nrows=1, ncols=1, sharex=False, sharey=False, squeeze=True, subplot_kw=None, gridspec_kw=None, **fig_kw)  
 <br>
 사용하는 parameter는 위와 같고,  
 <b>return값</b>이 <b>'fig'</b>와 <b>'ax'</b>로 두 개의 값을 가진다.  
 - <b>fig:</b> Figure
 - <b>ax:</b> Object or Array of Axes objects

```python
# First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

# Creates just a figure and only one subplot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')

# Creates two subplots and unpacks the output array immediately
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x, y)
ax1.set_title('Sharing Y axis')
ax2.scatter(x, y)
```

 <br>
 우선, figure와 하나의 subplot을 생성합니다.

```python
f, ax = plt.subplots(figsize = (5, 6))
```

## 3-1. Axis Grids
<br>
<b>sns.FacetGrid</b> (data, row=None, col=None, hue=None, col_wrap=None, sharex=True, sharey=True, height=3, aspect=1, palette=None, row_order=None, col_order=None, hue_order=None, hue_kws=None, dropna=True, legend_out=True, despine=True, margin_titles=False, xlim=None, ylim=None, subplot_kws=None, gridspec_kws=None, size=None)  
<br>
한 면에 grid를 나누는 방법으로, <b>data, row, col</b> 이면 충분하다.

```python
g = sns.FacetGrid(titanic, col="survived", row="sex")
```

grid를 위와 같이 변수에 세팅하고, <b>.map( )</b>을 통해 보고싶은 정보를 원하는 시각화 방법으로 입력한다.

```python
g = sns.FacetGrid(titanic, col="survived", row="sex")
g.map(plt.hist, "age")
```

```python
g = sns.FacetGrid(tips, col = "time", row = "smoker")
g.map(plt.scatter, "total_bill", "tip", s=9)
```

```python
kws = dict(s=50, linewidth=.5, edgecolor="w")
g = sns.FacetGrid(tips, col="sex", hue="time", palette="Set1", hue_order=["Dinner", "Lunch"])
g = (g.map(plt.scatter, "total_bill", "tip", **kws).add_legend())
```
